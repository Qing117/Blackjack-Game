```python
import random
```

```python
cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
number_of_cards = len(cards)
play = 'y'

# write a function to determine the card drawn, if Ace is drawn
def decide_draw(score, draw):
    if score > 10 and draw == 11:
        draw = 1
    return draw

playing = True
while playing:
    print("\n")
    play = input("Do you want to play Blackjack? Type 'y' or 'n':").lower()
    if play == 'n':
        playing = False
    else:
        your_score = 0
        computer_score = 0
        your_cards = []
        computer_cards = []

        # you draw two cards
        draw = cards[random.randint(0, number_of_cards - 1)]
        your_draw = decide_draw(your_score, draw)
        your_cards.append(your_draw)
        your_score = your_score + your_draw
        draw = cards[random.randint(0, number_of_cards - 1)]
        your_draw = decide_draw(your_score, draw)
        your_cards.append(your_draw)
        your_score = your_score + your_draw
        print(f"Your cards: {your_cards}, current score:{your_score}")

        # computer draw two cards and shows the first card
        draw = cards[random.randint(0, number_of_cards - 1)]
        computer_draw = decide_draw(computer_score, draw)
        computer_first_card = computer_draw
        computer_cards.append(computer_draw)
        computer_score = computer_score + computer_draw
        print(f"Computer's first card:{computer_first_card}")
        draw = cards[random.randint(0, number_of_cards - 1)]
        computer_draw = decide_draw(computer_score, draw)
        computer_cards.append(computer_draw)
        computer_score = computer_score + computer_draw

        # check if you win the blackjack

        if your_score == 21 and computer_score < 21:
            print(f"Your final hand: {your_cards}, final score: {your_score}")
            print(f"Computer's final hand: {computer_cards}, final score: {computer_score}")
            print("You win the Blackjack.")
        elif your_score == 21 and computer_score == 21:
            print(f"Your final hand: {your_cards}, final score: {your_score}")
            print(f"Computer's final hand: {computer_cards}, final score: {computer_score}")
            print("It is a draw")
        else:
            keep_play = True
            while your_score < 21 and keep_play:
                decision = input("Type 'y' to get another card, type 'n' to pass:")
                if decision == 'n':
                    print(f"Your final hand: {your_cards}, final score: {your_score}")
                    while computer_score < 17:
                        draw = cards[random.randint(0, number_of_cards - 1)]
                        computer_draw = decide_draw(computer_score, draw)
                        computer_cards.append(computer_draw)
                        computer_score = computer_score + computer_draw
                    print(f"Computer's final hand: {computer_cards}, final score: {computer_score}")
                    if computer_score > 21:
                        print("Computer went over. You win.")
                        keep_play = False
                    elif computer_score > your_score:
                        print("Computer win.")
                        keep_play = False
                    elif computer_score == your_score:
                        print("It is a draw")
                        keep_play = False
                    else:
                        print("You win.")
                        keep_play = False
                else:
                    draw = cards[random.randint(0, number_of_cards - 1)]
                    your_draw = decide_draw(your_score, draw)
                    your_cards.append(your_draw)
                    your_score = your_score + your_draw
                    print(f"Your cards: {your_cards}, current score: {your_score}")
                    print(f"Computer's first card:{computer_first_card}")
                    if your_score == 21 and computer_score < 21:
                        print(f"Your final hand: {your_cards}, final score: {your_score}")
                        print(f"Computer's final hand: {computer_cards}, final score: {computer_score}")
                        print("You win the Blackjack.")
                        keep_play = False
                    elif your_score == 21 and computer_score == 21:
                        print(f"Your final hand: {your_cards}, final score: {your_score}")
                        print(f"Computer's final hand: {computer_cards}, final score: {computer_score}")
                        print("It is a draw")
                        keep_play = False
                    elif your_score > 21 and computer_score <= 21:
                        print(f"Your final hand: {your_cards}, final score: {your_score}")
                        print(f"Computer's final hand: {computer_cards}, final score: {computer_score}")
                        print("You went over. Computer win")
                        keep_play = False
```
